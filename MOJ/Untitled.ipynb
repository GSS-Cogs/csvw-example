```python
import pandas as pd
from gssutils import pathify

df = pd.read_csv("./example1/probation-caseload_all-supervision.csv")

df.head()
```

```python

for col in ["Area"]:
    
    d = {
        "Label":[],
        "Notation":[],
        "Parent Notation":[],
        "Sort Priority":[],
        "Description":[]
    }
    for v in list(df[col].unique()):
        
        if col == "Sex":
            lookup = {"M": "Male", "F": "Female"}
            d["Label"].append(lookup[v])
            d["Notation"].append(v)
        else:
            d["Label"].append(v)
            d["Notation"].append(v)
        
        d["Parent Notation"].append("")
        d["Sort Priority"].append("")
        d["Description"].append("")
        
        pd.DataFrame.from_dict(d).to_csv(pathify(col)+".csv", index=False)
```
